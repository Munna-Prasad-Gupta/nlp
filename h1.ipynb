```python
!pip install deep-translator
!pip install openai-whisper
!pip install spacy

# Download spaCy's English model
!python -m spacy download en_core_web_sm

import whisper
from deep_translator import GoogleTranslator
import spacy

# Step 1: Transcribe the audio using Whisper
def transcribe_audio(file_path):
    # Load the Whisper model
    model = whisper.load_model('large')  # You can use 'base', 'medium', or 'large'

    # Perform transcription
    result = model.transcribe(file_path)

    # Return the transcribed text
    return result["text"]

# Step 2: Translate the transcription to English using deep_translator
def translate_text(text, target_language='en'):
    translated = GoogleTranslator(source='auto', target=target_language).translate(text)
    return translated

# Step 3: Perform POS tagging and Named Entity Recognition (NER) on the text
def pos_tagging_and_ner(text):
    # Load spaCy's English model
    nlp = spacy.load("en_core_web_sm")

    # Process the text through spaCy
    doc = nlp(text)

    # Extract POS tags and Named Entities
    pos_tags = [(word.text, word.pos_) for word in doc]
    named_entities = [(ent.text, ent.label_) for ent in doc.ents]

    return pos_tags, named_entities

# Main function to tie everything together
def process_audio(file_path):
    # Step 1: Transcribe the audio to text
    transcription = transcribe_audio(file_path)
    print(f"Original Transcription: {transcription}\n")

    # Step 2: Translate the transcription to English
    translated_text = translate_text(transcription, target_language='en')
    print(f"Translated to English: {translated_text}\n")

    # Step 3: Apply POS tagging and NER on the translated text
    pos_tags, named_entities = pos_tagging_and_ner(translated_text)

    print("POS Tagging Result:")
    for word, pos in pos_tags:
        print(f"{word}: {pos}")

    print("\nNamed Entity Recognition Result:")
    for entity, label in named_entities:
        print(f"{entity}: {label}")

    # Optionally, save the result to a file
    with open("translated_transcription_with_pos_and_ner.txt", "w") as file:
        file.write(f"Original Transcription:\n{transcription}\n\n")
        file.write(f"Translated Text (English):\n{translated_text}\n\n")
        file.write("POS Tagging:\n")
        for word, pos in pos_tags:
            file.write(f"{word}: {pos}\n")
        file.write("\nNamed Entity Recognition:\n")
        for entity, label in named_entities:
            file.write(f"{entity}: {label}\n")

# Run the process with your audio file path
file_path = 'common_voice_hi_41115996.mp3'  # Replace with the path to your audio file
process_audio(file_path)
```
